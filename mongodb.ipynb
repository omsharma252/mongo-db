### Q1. What are the key differences between SQL and NoSQL databases?

| Feature | SQL | NoSQL |
|---|---|---|
| Schema | Fixed, predefined schema | Dynamic, flexible schema |

SQL databases are best when data relationships are complex and consistency is critical. NoSQL databases are preferred when dealing with large volumes of unstructured or semi-structured data.

### Q3. Explain the concept of collections in MongoDB.

A **collection** in MongoDB is a group of documents stored together. It is analogous to a **table** in relational databases.

Each collection contains multiple BSON documents.

### Q4. How does MongoDB ensure high availability using replication?

MongoDB uses **Replica Sets** to ensure high availability.

A Replica Set consists of:
- **Primary Node**: Receives all write operations.
- **Secondary Nodes** (one or more): Replicate data from the primary node asynchronously.
- **Arbiter** (optional): Participates in elections but does not store data.

**How it ensures high availability:**
1. If the primary node fails, an **automatic election** is triggered among secondary nodes.
2. The node with the most up-to-date data and highest priority becomes the new primary.
3. The entire election process takes only a few seconds, minimizing downtime.
4. Clients automatically reconnect to the new primary.

**Minimum recommended setup:** 3-node replica set (1 primary + 2 secondaries).

### Q6. What is the role of indexes in MongoDB, and how do they improve performance?

**Indexes** in MongoDB are special data structures that store a portion of the collection's data in an easy-to-traverse form. They significantly speed up query execution.

**Without an index**: MongoDB performs a **Collection Scan** — it reads every document to find matches (O(n) time complexity).  
**With an index**: MongoDB performs an **Index Scan** — it traverses the index tree to find matches quickly (O(log n) time complexity).

**Types of Indexes in MongoDB:**
- **Single Field Index**: Index on one field.
- **Compound Index**: Index on multiple fields.
- **Multikey Index**: Index on array fields.
- **Text Index**: Supports full-text search.
- **Geospatial Index**: Supports location-based queries.
- **Hashed Index**: Used for hash-based sharding.

**Example:**
```python
collection.create_index([("Region", 1)])  # Ascending index on Region field
```

**Trade-off:** Indexes improve read performance but slightly slow down write operations and consume extra storage.

### Q8. What is sharding in MongoDB? How does it differ from replication?

**Sharding** is MongoDB's method for **horizontal scaling**. It distributes data across multiple servers (shards).

- A **shard key** is chosen to determine how data is distributed.
- Each shard holds a subset of the data.
- A **mongos** router directs queries to the appropriate shard(s).
- **Config servers** store metadata about the cluster.

**Sharding vs Replication:**

| Feature | Sharding | Replication |
|---|---|---|
| Purpose | Scalability (handles more data/load) | High Availability (fault tolerance) |
| Data | Each shard holds a **different** subset of data | Each replica holds the **same** copy of data |
| Benefit | Distributes storage and query load | Ensures data redundancy and failover |
| Use when | Dataset is too large for one server | Need zero-downtime and data backup |

In production, both are typically used together: a sharded cluster where each shard is itself a replica set.

### Q10. What are the ACID properties in the context of MongoDB transactions?

MongoDB (since version 4.0) supports multi-document **ACID transactions**.

| Property | Description | MongoDB Support |
|---|---|---|
| **Atomicity** | All operations in a transaction succeed or all are rolled back | ✅ Supported |
| **Consistency** | Database remains in a valid state before and after the transaction | ✅ Supported |
| **Isolation** | Concurrent transactions do not interfere with each other | ✅ Snapshot isolation |
| **Durability** | Committed transactions persist even after a system failure | ✅ Via journaling |

**Note:** For single-document operations, MongoDB has always been atomic. Multi-document transactions were introduced for more complex use cases like banking, e-commerce, etc.

### Q11. What is the purpose of MongoDB's explain() function?

The `explain()` function provides detailed information about how MongoDB **executes a query**. It is used for query performance analysis and optimization.

**It reveals:**
- Whether an index was used (IXSCAN) or a full collection scan was performed (COLLSCAN).
- Number of documents examined vs returned.
- Execution time in milliseconds.
- Query plan selected by the query optimizer.

**Usage:**
```python
collection.find({"Region": "West"}).explain("executionStats")
```

**Verbosity levels:**
- `queryPlanner` – Shows the winning plan.
- `executionStats` – Shows execution statistics.
- `allPlansExecution` – Shows stats for all candidate plans.

### Q12. How does MongoDB handle schema validation?

Although MongoDB is schema-less by default, it supports **schema validation** using **JSON Schema** to enforce rules on documents.

**Methods:**
1. **`$jsonSchema` validator** — Define rules at collection creation or modification.
2. **Validation levels**: `strict` (applies to all inserts/updates) or `moderate` (only for new documents).
3. **Validation actions**: `error` (reject invalid documents) or `warn` (log a warning).

**Example:**
```python
db.create_collection("orders", validator={
    "$jsonSchema": {
        "bsonType": "object",
        "required": ["OrderID", "Sales"],
        "properties": {
            "Sales": { "bsonType": "double", "minimum": 0 }
        }
    }
})
```

### Q13. What is the difference between a primary and a secondary node in a replica set?

| Feature | Primary Node | Secondary Node |
|---|---|---|
| **Role** | Handles all write operations | Replicates data from primary |
| **Reads** | Accepts reads by default | Can accept reads if `readPreference` is configured |
| **Writes** | All writes go here | Cannot accept writes |
| **Election** | Acts as current leader | Can become primary during an election |
| **Oplog** | Writes operations to oplog | Reads and replays the primary's oplog |
| **Count** | Only 1 primary at a time | Can have multiple secondaries |

If the primary node fails, the secondary nodes hold an election, and the most up-to-date secondary is promoted to the new primary.

### Q14. What security mechanisms does MongoDB provide for data protection?

MongoDB provides multiple layers of security:

1. **Authentication**: Supports SCRAM (default), x.509 certificates, LDAP, and Kerberos.
2. **Authorization (RBAC)**: Role-Based Access Control assigns specific permissions to users (e.g., `read`, `readWrite`, `dbAdmin`, `clusterAdmin`).
3. **TLS/SSL Encryption**: Encrypts data in transit between clients and the server.
4. **Encryption at Rest**: WiredTiger storage engine supports native encryption of stored data.
5. **Auditing**: Tracks and logs database activities for compliance.
6. **IP Whitelisting**: Restricts which IP addresses can connect to the database (especially in Atlas).
7. **Field-Level Encryption (FLE)**: Encrypts specific sensitive fields on the client side before sending to the server.
8. **Network Isolation**: VPC Peering and Private Endpoints in Atlas ensure network-level isolation.

### Q15. Explain the concept of embedded documents and when they should be used.

**Embedded documents** (also called nested documents) store related data within a single document rather than in a separate collection.

**Example:**
```json
{
  "_id": 1,
  "name": "John Doe",
  "address": {
    "street": "123 Main St",
    "city": "New York",
    "zip": "10001"
  }
}
```

**When to use embedded documents:**
- The related data is **always accessed together** with the parent document.
- There is a **one-to-one** or **one-to-few** relationship.
- The embedded data does **not grow unboundedly**.
- You want to avoid expensive joins (`$lookup`).

**When NOT to use (use references instead):**
- The related data is **frequently updated independently**.
- There is a **many-to-many** relationship.
- The nested data could grow very large (risk of exceeding 16MB BSON document limit).

### Q16. What is the purpose of MongoDB's $lookup stage in aggregation?

The `$lookup` stage performs a **left outer join** between two collections in the same database.

It allows combining documents from different collections — similar to SQL `JOIN`.

**Syntax:**
```python
{
  "$lookup": {
    "from": "<target_collection>",
    "localField": "<field_in_current_collection>",
    "foreignField": "<field_in_target_collection>",
    "as": "<output_array_field>"
  }
}
```

**Example:** Join `orders` collection with `customers` collection on `customer_id` to get customer details with each order.

The matched documents from the target collection are added as an **array field** in the output document.

### Q18. What are the advantages of using MongoDB for horizontal scaling?

**Horizontal scaling** means adding more servers to distribute the load, rather than upgrading a single server (vertical scaling).

MongoDB's advantages for horizontal scaling:

1. **Sharding**: Built-in native support to distribute data across multiple shards automatically.
2. **Shard Key Control**: Developers choose the shard key, allowing targeted queries to hit only relevant shards.
3. **Auto-Balancing**: The balancer process automatically redistributes chunks across shards.
4. **Cost-Effective**: Can use commodity hardware instead of expensive high-end servers.
5. **Elastic Scalability**: New shards can be added without downtime as data grows.
6. **Geographic Distribution**: Shards can be placed in different geographic regions for low-latency access.
7. **Parallel Processing**: Queries can be executed in parallel across multiple shards.

### Q19. How do MongoDB transactions differ from SQL transactions?

| Feature | MongoDB Transactions | SQL Transactions |
|---|---|---|
| **Scope** | Multi-document, multi-collection | Multi-row, multi-table |
| **Availability** | Available since MongoDB 4.0 | Always supported |
| **Performance** | Slight overhead; generally avoided for simple operations | Native and optimized |
| **Default Behavior** | Single-document operations are always atomic | All operations require explicit transaction handling |
| **Isolation Level** | Snapshot isolation | Multiple levels (READ COMMITTED, SERIALIZABLE, etc.) |
| **Use Case** | Required only for complex multi-document atomicity | Standard for most relational operations |

MongoDB recommends designing schemas so that multi-document transactions are **rarely needed**, as they add overhead compared to single-document operations.

### Q20. What are the main differences between capped collections and regular collections?

| Feature | Capped Collection | Regular Collection |
|---|---|---|
| **Size** | Fixed maximum size (in bytes) | Grows dynamically |
| **Document Limit** | Optional max document count | Unlimited |
| **Insertion Order** | Preserved (uses circular buffer) | Not guaranteed without sorting |

| **Update** | Allowed only if new document size = original size | No restriction |
| **Indexing** | `_id` index by default | Full indexing supported |
| **Use Cases** | Logging, caching, activity feeds | General-purpose storage |

### Q21. What is the purpose of the $match stage in MongoDB's aggregation pipeline?

The `$match` stage **filters documents** in the aggregation pipeline, passing only those that match the specified condition to the next stage.

It is equivalent to the `WHERE` clause in SQL.

**Best Practices:**
- Always place `$match` **as early as possible** in the pipeline to reduce the number of documents processed by subsequent stages.
- If placed as the first stage, MongoDB can use an **index** to optimize the query.

**Example:**
```python
pipeline = [
    { "$match": { "Region": "West", "Sales": { "$gt": 500 } } },
    { "$group": { "_id": "$Category", "totalSales": { "$sum": "$Sales" } } }
]
```
This filters only West region orders with Sales > 500, then groups them by Category.

### Q22. How can you secure access to a MongoDB database?

Steps to secure a MongoDB database:

1. **Enable Authentication**: Use `--auth` flag or set `security.authorization: enabled` in `mongod.conf`.
2. **Create User Accounts with Least Privilege**: Assign only the permissions needed.
   ```python
   db.create_user({ "user": "appUser", "pwd": "password", "roles": [{ "role": "readWrite", "db": "mydb" }] })
   ```
3. **Use TLS/SSL**: Encrypt all client-server communication.
4. **Bind to specific IP addresses**: Use `bindIp` to prevent unauthorized network access.
5. **Firewall Rules**: Block port 27017 from public access.
6. **Enable Auditing**: Monitor who accesses the database and what operations they perform.
7. **Regular Updates**: Keep MongoDB updated with the latest security patches.
8. **Use MongoDB Atlas**: Provides built-in security features including IP whitelisting, VPC peering, and encryption.

### Q23. What is MongoDB's WiredTiger storage engine, and why is it important?

**WiredTiger** is the default storage engine for MongoDB (since MongoDB 3.2).

**Key Features:**

1. **Document-Level Concurrency Control**: Uses optimistic concurrency control, allowing multiple clients to modify different documents simultaneously without locking the entire collection.
2. **Compression**: Supports Snappy (default), zlib, and zstd compression for collection and index data, reducing disk usage.
3. **Encryption at Rest**: Built-in support for AES256-CBC encryption (MongoDB Enterprise).
4. **Checkpointing**: Writes data to disk every 60 seconds, ensuring durability.
5. **Write-Ahead Logging (Journal)**: Ensures that operations are recorded before being applied, protecting against data loss on crashes.
6. **Better Performance**: Significantly outperforms the older MMAPv1 storage engine in both read and write-heavy workloads.

WiredTiger is important because it directly impacts MongoDB's **performance, scalability, and data durability**.
